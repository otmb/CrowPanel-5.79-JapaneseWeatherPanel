```python
import os
import glob
import convert

os.makedirs("weather2/output", exist_ok=True)
```

```python
size = 128
def svg_to_png(size):
    flist = glob.glob("weather2/*.svg")
    flist.sort()
    for fname in flist:
        name = os.path.basename(fname)
        name, _ = name.split(".")
        !rsvg-convert -b white -w {size} -h {size} --format=png --output=weather2/output/{name}_{size}.png {fname}

svg_to_png(128)
svg_to_png(64)
svg_to_png(32)
```

```python
!rm -f icons.py
def png_to_binary(size):
    flist = glob.glob(f"weather2/output/*_{size}.png")
    flist.sort()
    
    buffer_list = []
    for fname in flist:
        name = os.path.basename(fname)
        name, _ = name.split(".")
        name, _ = name.split("_")
        converted = convert.image_to_buffer(fname)
        buffer = converted[0]
    
        data = "'{}': {}".format(name, buffer)
        buffer_list.append(data)

    with open("icons.py", "a") as f:
        f.write(f"weather_icons{size}" + " = {\n  " + ",\n  ".join(buffer_list) + "\n}\n")

png_to_binary(128)
png_to_binary(64)
png_to_binary(32)
```
