```python
import json
with open("area.json") as f:
    data = json.load(f)
data
```

```python
data.keys()
```

```python
name_list = []
for code in data['centers']:
    for center_code in data['centers'][code]['children']:
        for office_code in data['offices'][center_code]['children']:
            for class10_code in data['class10s'][office_code]['children']:
                for class15_code in data['class15s'][class10_code]['children']:
                    print(
                        code, data['centers'][code]['name'],
                        data['centers'][code]['officeName'],
                        center_code, data['offices'][center_code]['name'],
                        data['offices'][center_code]['officeName'],
                        office_code, data['class10s'][office_code]['name'],
                        class10_code, data['class15s'][class10_code]['name'],
                        class15_code, data['class20s'][class15_code]['name'])
                    n = [
                        data['centers'][code]['name'],
                        data['centers'][code]['officeName'],
                        data['offices'][center_code]['name'],
                        data['offices'][center_code]['officeName'],
                        data['class10s'][office_code]['name'],
                        data['class15s'][class10_code]['name'],
                        data['class20s'][class15_code]['name']
                    ]
                    name_list.append("".join(n))
```

```python
name_list.append(" 0123456789/,℃°:-%().日月火水木金土年示エリア")
```

```python
texts = []

for name in name_list:
    for text in name:
        texts.append(text)

texts = list(set(texts))
texts = list(dict.fromkeys(texts))
text = "".join(texts)

text
```
